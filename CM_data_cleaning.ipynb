### Make Conviction Clusters

```python
cr = pd.read_csv("./base_data/conviction-rates.csv")
pr = pd.read_csv('./base_data/prosecution_all_years.csv')
```

```python
cr.head()
```

```python
pr.head()
```

```python
pr = pd.get_dummies(pr, 'Agency Name')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
pr_ss = StandardScaler().fit_transform(pr)
```

```python
inertia_list = []

for k in range(1, 51):
    kmeans = KMeans(n_clusters=k, random_state=12)
    kmeans.fit(pr_ss)
    inertia_list.append(kmeans.inertia_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_list = []

for k in range(2, 51):
    kmeans = KMeans(n_clusters=k, random_state=12)
    kmeans.fit(pr_ss)
    silhouette_list.append(silhouette_score(pr_ss, kmeans.labels_))
```

```python
silhouette_list_db = []

for ep in np.arange(.1,3,.1):
    dbscan = DBSCAN(eps=ep, n_jobs=-1)
    dbscan.fit(pr_ss)
    
    if len(np.unique(dbscan.labels_)) > 1:
        silhouette_list_db.append(silhouette_score(pr_ss, dbscan.labels_))
```

```python
plt.plot(inertia_list)
```

```python
plt.plot(silhouette_list)
```
